```python
plt.figure(figsize=(12,2))
plt.subplot(1,2,1)
plt.plot(x,np.cos(x))
plt.subplot(1,2,2)
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x,np.cos(x))
plt.subplot(1,2,2)
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.figure()
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
plt.subplot(2,2,1)
plt.plot(x,np.tan(x))
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='m')
plt.plot(x,np.sin(x-3),color=(1.0,0,0))
plt.show()
```
